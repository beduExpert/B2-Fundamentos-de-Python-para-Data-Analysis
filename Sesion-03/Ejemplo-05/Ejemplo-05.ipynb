#### Operaciones con NumPy Arrays.
NumPy permite operaciones entre arreglos y matrices de forma sencilla. 
la operación mas sencilla es la operación "punto a punto".

```python
import numpy as np
```

```python
x = np.array([[1,2],
              [2,1]])

y = np.array([[3,4],
              [4,3]])
```

```python
print(x + y)
print(x - y)
print(x * y)
print(x / y)
```
